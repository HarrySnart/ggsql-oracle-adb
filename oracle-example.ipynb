```python
import oracledb
from ggsqlOracle import ggsqlOracle
import pandas as pd
```

```python
# load connection details
config = pd.read_json('./config.json')
```

```python
# create oracledb connection
oracle_connection = oracledb.connect(user=config['user'][0], password=config['password'][0], dsn=config['dsn'][0],
                              config_dir=config['config_dir'][0],
                              wallet_location=config['wallet_dir'][0],wallet_password=config['wallet'][0])
```

```python
# create ggsqlOracle object direct from oracledb connection

ggsql = ggsqlOracle(oracle_connection)
```

```python
# fetch data
ggsql.sync(sql = ''' 
select * from HMEQ
''')
```

```python
# plot data
ggsql.plot(ggsql_query = ''' VISUALIZE reason as x, loan as y DRAW violin LABEL title => 'Violin Plot from Oracle Table' ''')
```

```python
# save plot
ggsql.save('./violin_plot.png')
```
