```ggsql
CREATE OR REPLACE VIEW penguins AS
SELECT *
FROM read_parquet('$HOME/ggsql/penguins.parquet')
```

```ggsql
SELECT *
FROM penguins
LIMIT 2
```

```ggsql
SELECT bill_len, bill_dep, species
FROM penguins
WHERE bill_len IS NOT NULL AND bill_dep IS NOT NULL

VISUALISE bill_len AS x, bill_dep AS y, species AS color
DRAW point
  SETTING opacity => 0.8
LABEL
  title => 'Penguin bill measurements',
  x => 'Bill length (mm)',
  y => 'Bill depth (mm)',
  color => 'Species'
```
