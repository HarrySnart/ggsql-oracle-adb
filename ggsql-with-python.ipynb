```python
import os
import polars as pl
import ggsql
```

```python
# Create a DataFrame
df = pl.DataFrame({
    "x": [1, 2, 3, 4, 5],
    "y": [10, 20, 15, 30, 25],
    "category": ["A", "B", "A", "B", "A"]
})
```

```python
# Render to Altair chart
plot = ggsql.render_altair(df, "VISUALIZE DRAW point MAPPING x as x, y as y, category as color LABEL title => 'Scatter plot using Python API'")
plot = plot.properties(width=500, height=320)
plot.display()
```
